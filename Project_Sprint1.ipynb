In the current iteration of our society, we are facing an unprecedented level of cubbyholing. Every person is in danger of being sliced up and put into neat little demographic slices by information servers. These can be advertisers, search engines, news agencies, governments or more.  So, how do we extract useful information out of measuements about subjective matters?  

<br>

I wanted to try this with polls about the current POTUS.  This is because he is very polarizing, and so much information has been generated.  Because of this bevy of information, you are unable to be shown all of it, and what you are shown is likely to influence your opinion.  This opinion may, therefore, not reflect reality as much as someones curated information or agenda.  This means it is hard to take any one poll seriously.   

<br>

To this end I have created a few charts which use a large sampling of polls to determine a more representative sample.  The first shows us the tracking of the POTUS' approval rating over time, with the dot size representing the samplesize.

<br>

<br>

As you can see, the larger the sample size, the more clustered the results.  By comparison, we can see all of the extreme outliers, for approve or disapprove, have very small sample sizes.  This shows something we all know: a smaller sample size does not necessarily represent the main trend.  It also shows that Trump's general approval rating has gone up over time.  A second chart showshis approval and disapproval, tracked since the last quarter of last year:

<br>

<br>

This shows us that his disapproval(in purple) has gone down by about 5 points since last year, and his approval(in olive) has gone up by 5 points since last year.  One interesting note is that the two colors are clearly demarcated, indicating that the two sides have pretty much made up their mind no matter what he does(with a small overlap for things like the announcement of the DPRK summit and the 1 year anniversary of Charlottesville).

these two charts give us a clearer picture of how the President is perceived across the entire population, and in case you are wondering how he stacks up to President Obama's record, the answer is surprising:

<br>
 
<br>

As you can see, Obama lost about 20 points by this time, whilst Trump has stayed steady and even made some small gains,  even through his extremely polarizing rhetoric.  This suggests that because he says so many things, people only hear the ones presented to them, and those are informed by your group, reinforcing your bias instead of allowing you to make a proper judgement.

##Visualizing Data

```
source = df538[5000:]

l = alt.Chart(source).mark_circle(size = 4, color = 'olive').encode(
    x=alt.X('enddate', axis=alt.Axis(title='Month (2018)')),
    y =alt.Y('mean(Approval Rating %)', axis=alt.Axis(title='Percentage Points', format = '%'), scale = alt.Scale(domain = [.3, .7])),    
    tooltip= ['end_date', 'adjusted_approve', 'pollster'],
).interactive().properties(
title = 'Approval has Increased, and Disapproval Decreased Over Time')




l2 = alt.Chart(source).mark_circle(size = 4, color = 'DarkSlateBlue').encode(
    x= 'enddate', 
    y='mean(Dispproval Rating %)',
    tooltip= ['end_date', 'adjusted_disapprove', 'pollster'],
).interactive()




v= l + l2
v
```

```
 source = df538[5000:]

c = alt.Chart(source).mark_circle(opacity = .45).encode(
    alt.X('startdate', scale=alt.Scale(zero=False), axis=alt.Axis(title='Month')), 
    y =alt.Y('mean(Approval Rating %)', axis=alt.Axis(title='Percentage Points', format = '%'), scale=alt.Scale(domain=[.32, .5])),
    tooltip=['adjusted_approve', 'adjusted_disapprove', 'pollster', 'end_date'],
    size = 'Population Sample Size'
).interactive().properties(
title = 'Outliers are Explained by Small Sample Size')
c 
```
